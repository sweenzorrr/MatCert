```python
import numpy as np
import pandas as pd
import plotly
import plotly.graph_objs as go

plotly.offline.init_notebook_mode(connected=True)

# Scientific libraries
import numpy as np
from scipy.optimize import curve_fit

PCTG08raw = pd.read_csv('Test.csv', delimiter = ',') #read the CSV raw data file and port into dataset

dfA = pd.DataFrame(
    {'E_Speed':[4,6,8,10,12],
     'Lower_limit':[65,410,770,1125,1485],
     'Upper_limit':[350,705,1065,1420,1780]})

A = [[4, 65, 350], 
    [6, 410, 705],
    [8, 770, 1065],
    [10, 1125, 1420],
    [12, 1485, 1780],
    [14, 1845, 2140],
    [16, 2200, 2490],
    [18, 2560, 2855],
    [20, 2910, 3185],
    [22, 3240, 3485],
    [24, 3540, 3765],
    [26, 3820, 4030],
    [28, 4075, 4275],
    [30, 4320, 4510],]

PCTG={} #dict
xdat={} #dict
ydat={} #dict
data = [] #list
coeffs = [] #list

for i in range(len(A)):
    PCTG["{}mms".format(2*(i+2))] = PCTG08raw.iloc[A[i][1]:A[i][2]]
    xdat["{}mms".format(2*(i+2))] = PCTG["{}mms".format(2*(i+2))].H_TEMP
    ydat["{}mms".format(2*(i+2))] = PCTG["{}mms".format(2*(i+2))].E_CURR

def exponenial_func(x_mms, a, b, c): #define an exponential function to fit the data
    return a+b*np.exp(c*x_mms)

# LOOPING OVER ALL THE DATA SETS IN PCTG08RAW DATAFRAME AND PLOTTING SCATTER AND FIT FUNCTIONS FOR EACH
for q in xdat.keys():
    x_vals = xdat[q]
    y_vals = ydat[q]
    
    popt, pcov = curve_fit(exponenial_func, x_vals, y_vals, p0=(1, 1, -1e-6))
    print("{} fit {}".format(q,popt))
    xx = np.linspace(200, 400, 1000)
    yy = exponenial_func(xx, *popt)
    coeffs.append(popt)
    pts = go.Scatter(
                  x=x_vals,
                  y=y_vals,
                  mode='markers',
                  #marker=go.Marker(color='rgb(255, 127, 14)'),
                  name=q
                  )
    data.append(pts)
    
    exp_fit = go.Scatter(
                  x=xx,
                  y=yy,
                  mode='lines',
                  #marker=go.Marker(color='rgb(255, 127, 14)'),
                  name=q+' fit'
                  )
    data.append(exp_fit)
    
    
layout = go.Layout(
                title='Exponential Fit in Python',
                plot_bgcolor='rgb(255, 255, 255)',
                  xaxis=go.layout.XAxis(zerolinecolor='rgb(220,220,220)', gridcolor='rgb(220,220,220)'),
                  yaxis=go.layout.YAxis(zerolinecolor='rgb(220,220,220)', gridcolor='rgb(220,220,220)'),
                )

fig = go.Figure(data=data, layout=layout)
fig
```
