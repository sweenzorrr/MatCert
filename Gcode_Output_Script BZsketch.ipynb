```python
import pandas as pd

df = pd.DataFrame(
{"E_speed" : [4,6,8,10], #E speed
"Temp" : [220,240,260,280]}) #Jamming temp for given E speed
#Type in material type
mat = 'PCTG' 
#Type in hozzle diameter
hoz = '0.4mm Hozzle' 

numt = 17 #number of temperatures to test for each velocity
timeint = 5 #time interval material should be extruded
tempint = 5 #temp interval change
f= open("Gcode_Output_Test.txt","w+")
f.write("; Begin Custom Script\r\n; ")
f.write(hoz)
f.write("\r\n; ")
f.write(mat)
f.write("\r\nTransMode13 \r\nTransAccel100 \r\nTransParam165 \r\n")
f.write("G92 E0 ;set extruder position to 0 \r\nM106 S0 ; turn off fan \r\n")

for i in range(len(df)):
    f.write(";Temp curve for {}mm/s\nM109 S{}\nG1 E{} F{}\nG92 E0\n"
            .format((df['E_speed'][i]),((tempint*numt)+(df['Temp'][i])),(timeint*(df['E_speed'][i])),(60*(df['E_speed'][i]))))
    for j in range(numt):
        f.write("M104 S{}\nG1 E{} F{}\nG92 E0\n"
               .format((((tempint*numt)+(df['Temp'][i])-((j+1)*tempint))),(timeint*(df['E_speed'][i])),(60*(df['E_speed'][i]))))
f.write(";ending Script\nM109 S220 T0")

f=open("Gcode_Output_Test.txt")   #creates an object file_o to access the file
print(f.read())                 #print-out the contents of file

f.close()
```
